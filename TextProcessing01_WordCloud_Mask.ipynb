```
# Libraries
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image    # to import the image

# for retrieving an image from URL
import requests
from io import BytesIO
 
# Create a list of word (https://en.wikipedia.org/wiki/Data_visualization)
#text="ALVAREZ, ARCE, CANDITA, CERVANTES, CHAUQUITA, MIGUEL, DIEGO, JOSE, GALDO, ILLAPU, WILMER, EMILIANO, JOYO, MATEO, PEÑA, MUNARRIZ, NAVARRETE, SEBASTIAN, SAMIR, JHOSHI, ROSCA, DAVID, ROJITAS, STEPHEN, JUANCHO, EDDY, TORRE, VELARDE, VILA, YANCE, ANTONIO, ALESSANDRO, AVILA, CHAVEZ, ESTEBAN, DIEGO, DODI, ESPINO, KALEF, FALCO, GABRIEL, RITAS,  CIPRIANO, MARTINEZ, MATIAS, MUNAYLLA, NORIEGA, ROMEL, MAURI, CHOCHO, ESTEBAN, GONZO, RALPH, SOLES, VALLEJO, VECTOR, DANIEL, ZEVALLOS, CANTU, CANTU, DOMINGO, DOMINGO, DOMINGO, VECTOR"
 
# Load the image locally (from your PC)
response = requests.get('https://raw.githubusercontent.com/CVilaNet/Visualizacion/main/img/DonBosco_01.jpg')
mymask = np.array(Image.open(BytesIO(response.content)))
 
# Make the figure
wordcloud = WordCloud(mask=mymask, background_color="white", contour_width=0.1).generate(text)

plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
